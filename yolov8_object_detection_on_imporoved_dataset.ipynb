```python
import tensorflow as tf
import cv2
import numpy as np
import os

# Path to the exported SavedModel directory from YOLOv8
saved_model_dir = "path_to_saved_model"  # 👈 Replace with actual path (e.g., 'runs/train/exp/weights/best_saved_model')

# Path to folder with real images for quantization (use 100 sample pothole images)
representative_data_dir = "/content/road-damage-4/valid"  # 👈 Replace with your folder path (e.g., 'data/images/val')

# Define the representative dataset function
def representative_dataset():
    image_paths = [os.path.join(representative_data_dir, fname)
                   for fname in os.listdir(representative_data_dir) if fname.endswith(".jpg")]

    for path in image_paths[:100]:  # You can increase this number if needed
        img = cv2.imread(path)
        img = cv2.resize(img, (640, 640))  # Adjust to your model input size
        img = img.astype(np.float32)
        img = np.expand_dims(img, axis=0)  # Add batch dimension
        yield [img]

# Create the converter
converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_dir)

# Enable full integer quantization
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_dataset
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]

# Set input/output type to uint8 (common for mobile deployment)
converter.inference_input_type = tf.uint8
converter.inference_output_type = tf.uint8

# Convert the model
tflite_quant_model = converter.convert()

# Save the quantized model
output_path = "yolov8_int8_quantized.tflite"
with open(output_path, "wb") as f:
    f.write(tflite_quant_model)

print(f"✅ INT8 Quantized model saved to: {output_path}")
```
